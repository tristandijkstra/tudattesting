```python
import sys
import datetime
import matplotlib.pyplot as plt
sys.path.insert(0, r"/mnt/c/Users/Trez/Desktop/tudat-bundle/tudatpy/")
from tudatpy.data.horizons import HorizonsQuery, HorizonsBatch
import time
import matplotlib.pyplot as plt
import numpy as np
```

```python
lst = [f"{x};" for x in range(1, 20)] + ["399", "499", "599"]

start_dt = datetime.datetime(2020, 1, 1)
end_dt = datetime.datetime(2023, 6, 1)
```

```python
batch = HorizonsBatch(
    query_id_list=lst,
    location="@SSB",
    epoch_start=start_dt,
    epoch_end=end_dt,
    epoch_step="15d",
    extended_query=True
)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (9, 9))

for obj in batch._query_objects.values():
    vec = obj.carthesian()
    ax.plot(vec[:, 1], vec[:, 2], linewidth="0.8")
    ax.scatter(vec[-1, 1], vec[-1, 2])

ax.grid()
ax.set_aspect("equal")

plt.show()
```
